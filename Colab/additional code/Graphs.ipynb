```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classes_grouped = [['YOLOv4-tiny 3 classes', 'YOLOv4-tiny 2 classes', 'YOLOv4(full) 3classes', 'YOLOv4(full) 2classes'], [47.54, 32.84, 96.01, 97.38], [45.70, 26.36, 95.10, 96.17], [35.76, 0.00, 95.09, 0.00]]
classes_names = ['Methods', 'Correct Face Mask Wearing', 'No Face Mask Wearing', 'Incorrect Face Mask Wearing']
```

```python
df = pd.DataFrame(classes_grouped, index=classes_names).T
df
```

```python
values = [47.54, 45.70, 35.76] + [32.84, 26.36, 0.00] + [96.01, 95.1, 95.09] + [97.38, 96.17, 0.0]
labels = ['47.54 %', '45.70 %', '35.76 %', '32.84 %', '26.36 %', '', '96.01%', '95.10%', '95.09%', '97.38%', '96.17%', '']
rows = [-0.25, -0.09, 0.075, 0.75, 0.91, 1.10, 1.75, 1.92, 2.09, 2.75, 2.92, 0]

df.plot(x='Methods',
        y=['Correct Face Mask Wearing', 'No Face Mask Wearing', 'Incorrect Face Mask Wearing'],
        kind='bar',
        figsize=(20,8),
        colormap='copper',
        rot=0)

plt.ylabel('Average Accuracy', fontsize=20)
plt.xlabel('Methods', fontsize=20)

for i in range(len(values)):
    plt.text(x = rows[i] , y = values[i]-4, s = labels[i], size = 12, c='white')
```
